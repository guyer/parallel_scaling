# Dendrite derivation

\begin{align}
\tau(\hat{n}) \frac{\partial \phi}{\partial t}
&= [\phi - \lambda U(1 - \phi^2)](1 - \phi^2)
+ \nabla\cdot\left([W(\hat{n})]^2 \nabla \phi\right)
+ \frac{\partial}{\partial x}\left[\lvert\nabla\phi\rvert^2 W(\hat{n})
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial x}\right)}\right]
+ \frac{\partial}{\partial y}\left[\lvert\nabla\phi\rvert^2 W(\hat{n})
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial y}\right)}\right]
\\
W(\hat{n}) &= W_0 \left\{1 + \epsilon_m \cos\left[m(\theta - \theta_0)\right]\right\}
\end{align}

\begin{align}
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial x}\right)}
&= W_0 \epsilon_m \frac{\partial \cos\left[m(\theta - \theta_0)\right]}{\partial\left(\frac{\partial\phi}{\partial x}\right)}
\\
&= - W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m \frac{\partial \theta}{\partial\left(\frac{\partial\phi}{\partial x}\right)}
\end{align}

\begin{align}
\tan \theta &= n_y / n_x
\\
&= \frac{\frac{\partial\phi}{\partial y}}{\lvert\nabla\phi\rvert}
\frac{\lvert\nabla\phi\rvert}{\frac{\partial\phi}{\partial x}}
\\
&= \frac{\frac{\partial\phi}{\partial y}}{\frac{\partial\phi}{\partial x}}
\\
&\stackrel{?}{=} \frac{\sin\theta}{\cos\theta}
\end{align}

\begin{align}
\frac{\partial \tan \theta}{\partial\left(\frac{\partial\phi}{\partial x}\right)} 
&= \frac{\partial}{\partial\left(\frac{\partial\phi}{\partial x}\right)} \frac{\frac{\partial\phi}{\partial y}}{\frac{\partial\phi}{\partial x}}
\\
\frac{1}{\cos^2\theta} \frac{\partial \theta}{\partial\left(\frac{\partial\phi}{\partial x}\right)} 
&= -\frac{\frac{\partial\phi}{\partial y}}{\left(\frac{\partial\phi}{\partial x}\right)^2}
\\
\frac{\partial \theta}{\partial\left(\frac{\partial\phi}{\partial x}\right)} 
&= -\frac{\sin\theta \cos\theta}{\frac{\partial\phi}{\partial x}}
\end{align}

\begin{align}
\frac{\partial \tan \theta}{\partial\left(\frac{\partial\phi}{\partial y}\right)} 
&= \frac{\partial}{\partial\left(\frac{\partial\phi}{\partial y}\right)} \frac{\frac{\partial\phi}{\partial y}}{\frac{\partial\phi}{\partial x}}
\\
\frac{1}{\cos^2\theta} \frac{\partial \theta}{\partial\left(\frac{\partial\phi}{\partial y}\right)} 
&= \frac{1}{\frac{\partial\phi}{\partial x}}
\\
\frac{\partial \theta}{\partial\left(\frac{\partial\phi}{\partial y}\right)} 
&= \frac{\cos^2\theta}{\frac{\partial\phi}{\partial x}}
\end{align}

\begin{align}
\frac{\partial \tan \theta}{\partial x} 
&= \frac{\partial}{\partial x} \frac{\frac{\partial\phi}{\partial y}}{\frac{\partial\phi}{\partial x}}
\\
\frac{1}{\cos^2\theta} \frac{\partial \theta}{\partial x} 
&= 
\frac{\frac{\partial^2\phi}{\partial x \partial y}}{\frac{\partial\phi}{\partial x}}
- \frac{\frac{\partial\phi}{\partial y}}{\left(\frac{\partial\phi}{\partial x}\right)^2}
\frac{\partial^2\phi}{\partial x^2}
\\
\frac{\partial \theta}{\partial x} 
&= 
\cos^2\theta
\left[\frac{\frac{\partial^2\phi}{\partial x \partial y}}{\frac{\partial\phi}{\partial x}}
- \frac{\frac{\partial\phi}{\partial y}}{\left(\frac{\partial\phi}{\partial x}\right)^2}
\frac{\partial^2\phi}{\partial x^2}
\right]
\\
&= 
\cos^2\theta
\left[
\frac{\frac{\partial\phi}{\partial x}}{\left(\frac{\partial\phi}{\partial x}\right)^2}
\frac{\partial^2\phi}{\partial x \partial y}
- \frac{\frac{\partial\phi}{\partial y}}{\left(\frac{\partial\phi}{\partial x}\right)^2}
\frac{\partial^2\phi}{\partial x^2}
\right]
\end{align}

\begin{align}
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial x}\right)}
&= - W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m \frac{\partial \theta}{\partial\left(\frac{\partial\phi}{\partial x}\right)}
\\
&= W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m \frac{\sin\theta \cos\theta}{\frac{\partial\phi}{\partial x}}
\\
&= W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m \sin\theta
\end{align}

\begin{align}
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial y}\right)}
&= - W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m \frac{\partial \theta}{\partial\left(\frac{\partial\phi}{\partial y}\right)}
\\
&= -W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m \cos\theta
\end{align}

\begin{align}
\lvert\nabla\phi\rvert^2 W(\hat{n})
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial x}\right)}
&= \lvert\nabla\phi\rvert^2 W(\hat{n})
W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m \sin\theta
\\
&= \lvert\nabla\phi\rvert^2 W_0^2 \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\epsilon_m \sin\left[m(\theta - \theta_0)\right] m \sin\theta
\end{align}

\begin{align}
\frac{\partial}{\partial x}\left[\lvert\nabla\phi\rvert^2 W(\hat{n})
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial x}\right)}\right]
&= W_0^2 \epsilon_m m \frac{\partial}{\partial x}
\left(
    \lvert\nabla\phi\rvert^2 \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\sin\left[m(\theta - \theta_0)\right] \sin\theta
\right)
\\
&= W_0^2 \epsilon_m m 
\left(
    2 \lvert\nabla\phi\rvert \frac{\partial\lvert\nabla\phi\rvert}{\partial x} \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\sin\left[m(\theta - \theta_0)\right] \sin\theta
\right.
\\
&\qquad\qquad\qquad\left.
- \lvert\nabla\phi\rvert^2 
\epsilon_m\sin^2\left[m(\theta - \theta_0)\right]\sin\theta \, m\frac{\partial\theta}{\partial x}
\right.
\\
&\qquad\qquad\qquad\left.
+ \lvert\nabla\phi\rvert^2 \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\cos\left[m(\theta - \theta_0)\right] \sin\theta
\, m\frac{\partial\theta}{\partial x}
\right.
\\
&\qquad\qquad\qquad\left.
+\lvert\nabla\phi\rvert^2 \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\sin\left[m(\theta - \theta_0)\right] \cos\theta \frac{\partial\theta}{\partial x}
\right)
\end{align}

\begin{align}
\frac{\partial\lvert\nabla\phi\rvert}{\partial x}
&=
\frac{\partial\left[\left(\nabla\phi\right)^2\right]^{1/2}}{\partial x}
\\
&=
\frac{1}{2}\left[\left(\nabla\phi\right)^2\right]^{-1/2}\frac{\partial\left(\nabla\phi\right)^2}{\partial x}
\\
&=
\frac{1}{2}\left[\left(\nabla\phi\right)^2\right]^{-1/2}2\nabla\phi\frac{\partial\nabla\phi}{\partial x}
\\
&=
\frac{\nabla\phi}{\lvert\nabla\phi\rvert}\frac{\partial\nabla\phi}{\partial x}
\\
&=
\hat{n}\frac{\partial\nabla\phi}{\partial x}
\end{align}

\begin{align}
\frac{\partial}{\partial x}\left[\lvert\nabla\phi\rvert^2 W(\hat{n})
\frac{\partial W(\hat{n})}{\partial\left(\frac{\partial\phi}{\partial x}\right)}\right]
&= W_0^2 \epsilon_m m 
\left(
    2 \nabla\phi\frac{\partial\nabla\phi}{\partial x} \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\sin\left[m(\theta - \theta_0)\right] \sin\theta
\right.
\\
&\qquad\qquad\qquad\left.
- \lvert\nabla\phi\rvert^2 
\epsilon_m\sin^2\left[m(\theta - \theta_0)\right]\sin\theta \, m\frac{\partial\theta}{\partial x}
\right.
\\
&\qquad\qquad\qquad\left.
+ \lvert\nabla\phi\rvert^2 \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\cos\left[m(\theta - \theta_0)\right] \sin\theta
\, m\frac{\partial\theta}{\partial x}
\right.
\\
&\qquad\qquad\qquad\left.
+\lvert\nabla\phi\rvert^2 \left\{
1 + \epsilon_m\cos\left[m(\theta - \theta_0)\right]
\right\}
\sin\left[m(\theta - \theta_0)\right] \cos\theta \frac{\partial\theta}{\partial x}
\right)
\end{align}

\begin{align}
\frac{\delta \mathcal{F}}{\delta \phi}
&= \frac{\delta}{\delta \phi}\left\{
\frac{1}{2}\left[
W(\hat{n})
\right]^2
\lvert\nabla\phi\rvert^2
\right\}
\\
&= \frac{\partial}{\partial \phi}\left\{
\frac{1}{2}\left[
W(\hat{n})
\right]^2
\lvert\nabla\phi\rvert^2

- \nabla\cdot\frac{\partial}{\partial\nabla\phi}
\left\{
\frac{1}{2}\left[
W(\hat{n})
\right]^2
\lvert\nabla\phi\rvert^2
\right\}
\\
&= - \nabla\cdot
\left\{
W(\hat{n})

+ \left[
W(\hat{n})
\right]^2
\lvert\nabla\phi\rvert

\right\}

&= - \nabla\cdot
\left\{
W(\hat{n})

W(\hat{n})
\right]^2
\nabla\phi

\end{align}

\begin{align}
\frac{\partial W(\hat{n})}{\partial\nabla\phi}
&= -W_0 \epsilon_m \sin\left[m(\theta - \theta_0)\right] m
\frac{\partial\theta}{\partial\nabla\phi}
\end{align}

\begin{align}
\tan\theta
&= \frac{\frac{\partial\phi}{\partial y}}{\frac{\partial\phi}{\partial x}}
\\
\frac{\partial}{\partial\nabla\phi}\tan\theta
&= \frac{\partial}{\partial\nabla\phi}\left(
\frac{\frac{\partial\phi}{\partial y}}{\frac{\partial\phi}{\partial x}}
\right)
\\
\frac{1}{\cos^2\theta}\frac{\partial\theta}{\partial\nabla\phi}
&= \frac{\partial}{\partial\nabla\phi}\left(
\frac{\nabla\phi\cdot\hat{j}}{\nabla\phi\cdot\hat{i}}
\right)
\\
\frac{1}{\cos^2\theta}\frac{\partial\theta}{\partial\nabla\phi}
&= 
\frac{\hat{j}}{\nabla\phi\cdot\hat{i}}
- \frac{\nabla\phi\cdot\hat{j}}{\left(\nabla\phi\cdot\hat{i}\right)^2} \hat{i}
\\
&= 
\frac{\nabla\phi\cdot\hat{i}}{\left(\nabla\phi\cdot\hat{i}\right)^2}\hat{j}
- \frac{\nabla\phi\cdot\hat{j}}{\left(\nabla\phi\cdot\hat{i}\right)^2} \hat{i}
\\
\frac{\lvert\nabla\phi\rvert^2}{\left(\frac{\partial\phi}{\partial x}\right)^2}\frac{\partial\theta}{\partial\nabla\phi}
&= 
\frac{\frac{\partial\phi}{\partial x}}
{\left(\frac{\partial\phi}{\partial x}\right)^2}\hat{j}
- \frac{\frac{\partial\phi}{\partial y}}
{\left(\frac{\partial\phi}{\partial x}\right)^2} \hat{i}
\\
\frac{\partial\theta}{\partial\nabla\phi}
&= 
\frac{1}{\lvert\nabla\phi\rvert^2}
\left(\frac{\partial\phi}{\partial x} \hat{j}
- \frac{\partial\phi}{\partial y} \hat{i}\right)
\end{align}
